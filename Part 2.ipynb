2. Describe a practical experiment you would design to compare the effectiveness of the
proposed change in relation to the key measure of success. Please provide details on:
a. how you will implement the experiment
b. what statistical test(s) you will conduct to verify the significance of the
observation
c. how you would interpret the results and provide recommendations to the city
operations team along with any caveats.


a) For this experiment, I would use weekends in order to test the effect of paying for tolls, reason being that this is a time when the usage is more similar between the cities. I would offer 50% of the rideshare providers in each city access to the cash reimbursement program so that we can measure the response among population from each. I would track the following:
- The amount of time the providers spent at both cities. 
- Wait time of the clients
- The cost per mile of the subsequent rides provided. 
- The revenue generated by the providers during the period. 
- Establish that H0, the null hypothesis is that the providers don't spend more time in each city even with the program in place.

b) the statistical tests I would employ to verify impact would be to check the p-value of the wait times, trip cost per mile.

c) In order to guide the product team from these results I would share the insights regarding how the A/B test impacted the behavior of the providers. If it turns out that the toll reimbursement ends up driving a statistically significant impact to the the measure metrics, we could discuss options regarding how we communicate this to the providers so that they are more encouraged to pay the toll to pick up work or consider payin this fee for them.
